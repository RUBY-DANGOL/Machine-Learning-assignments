```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 3, 4, 5])         # Independent variable (Hours studied)
y = np.array([50, 55, 65, 70, 75])    # Dependent variable (Exam scores)

# Calculate means
x_mean = np.mean(X)
y_mean = np.mean(y)

# Calculate slope (β1) and intercept (β0)
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
slope = numerator / denominator
intercept = y_mean - slope * x_mean

# Predict values
y_pred = slope * X + intercept

# Plot
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Simple Linear Regression (NumPy)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset: 2 features (e.g., hours studied and classes attended)
# Format: [feature1, feature2]
X = np.array([
    [1, 4],
    [2, 3],
    [3, 5],
    [4, 2],
    [5, 1]
])

# Target variable: exam scores
y = np.array([52, 55, 65, 70, 75])

# Add bias (intercept) term to X (column of ones)
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # shape: (n_samples, n_features+1)

# Compute parameters (θ) using Normal Equation
theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

# Predict target values
y_pred = X_b @ theta

# Print coefficients
print("Intercept (θ₀):", theta[0])
print("Coefficients (θ₁, θ₂):", theta[1:])

# Plot actual vs predicted values
plt.scatter(range(len(y)), y, color='blue', label='Actual Scores')
plt.plot(range(len(y_pred)), y_pred, color='red', label='Predicted Scores')
plt.xlabel('Sample Index')
plt.ylabel('Exam Score')
plt.title('Multiple Linear Regression (NumPy)')
plt.legend()
plt.show()
```
